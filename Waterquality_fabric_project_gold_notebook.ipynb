```sql
-- Creatig Yearly Lead Summary per Location (Group lead readings per year and location)

CREATE OR REPLACE TABLE Gold_Lead_Yearly_Summary AS
SELECT
  location_id,
  YEAR(TO_DATE(sample_date, 'MM-dd-yyyy')) AS year,
  COUNT(*) AS total_samples,
  AVG(lead_ppb) AS avg_lead_ppb,
  MAX(lead_ppb) AS max_lead_ppb,
  MIN(lead_ppb) AS min_lead_ppb
FROM Silver_Lead
WHERE lead_ppb IS NOT NULL
GROUP BY location_id, YEAR(TO_DATE(sample_date, 'MM-dd-yyyy'))

```

```sql
--Creating gold layer table for Flag Locations with Consistent Unsafe Readings (Flag locations where more than 3 samples in a year exceed 15 ppb (unsafe limit))

CREATE OR REPLACE TABLE Gold_Lead_Unsafe_Locations AS
SELECT
  location_id,
  YEAR(TO_DATE(sample_date, 'MM-dd-yyyy')) AS year,
  COUNT(*) AS unsafe_sample_count
FROM Silver_Lead
WHERE lead_ppb > 15
GROUP BY location_id, YEAR(TO_DATE(sample_date, 'MM-dd-yyyy'))
HAVING COUNT(*) >= 3
```

```sql
--Creating table for Monthly Average Lead Level per Location (Useful for time-series tracking)

CREATE OR REPLACE TABLE Gold_Lead_Monthly_Avg AS
SELECT
  location_id,
  YEAR(TO_DATE(sample_date, 'MM-dd-yyyy')) AS year,
  MONTH(TO_DATE(sample_date, 'MM-dd-yyyy')) AS month,
  AVG(lead_ppb) AS avg_monthly_lead_ppb
FROM Silver_Lead
WHERE lead_ppb IS NOT NULL
GROUP BY location_id, YEAR(TO_DATE(sample_date, 'MM-dd-yyyy')), MONTH(TO_DATE(sample_date, 'MM-dd-yyyy'))
```

```sql
--Creating table for Exceedance Events (>15 µg/L)(Each time a sample exceeds 15 µg/L = "exceedance event")

CREATE OR REPLACE TABLE Gold_Lead_Exceedance_Events AS
SELECT
  location_id,
  sample_date,
  lead_ppb,
  result_status,
  detection_condition,
  collection_method,
  analysis_date
FROM Silver_Lead
WHERE lead_ppb > 15
```

```sql
--Creating table for Completeness Report (it will Help assess if there are gaps in monthly data submission)

CREATE OR REPLACE TABLE Gold_Lead_Completeness_Report AS
SELECT
  location_id,
  YEAR(TO_DATE(sample_date, 'MM-dd-yyyy')) AS year,
  MONTH(TO_DATE(sample_date, 'MM-dd-yyyy')) AS month,
  COUNT(*) AS samples_count
FROM Silver_Lead
GROUP BY location_id, YEAR(TO_DATE(sample_date, 'MM-dd-yyyy')), MONTH(TO_DATE(sample_date, 'MM-dd-yyyy'))
```

```sql
-- Creating Table for Lead Risk Summary (Based on lead_risk_level column)

CREATE OR REPLACE TABLE Gold_Lead_Risk_Summary AS
SELECT
  location_id,
  lead_risk_level,
  COUNT(*) AS sample_count,
  AVG(lead_ppb) AS avg_lead,
  MAX(lead_ppb) AS max_lead
FROM Silver_Lead
WHERE lead_ppb IS NOT NULL
GROUP BY location_id, lead_risk_level
```

```sql
-- Saveing  All Gold Tables in Delta Format (Save each Gold table as Delta)


CREATE OR REPLACE TABLE Gold_Lead_Yearly_Summary
USING delta
AS SELECT * FROM Gold_Lead_Yearly_Summary;

CREATE OR REPLACE TABLE Gold_Lead_Unsafe_Locations
USING delta
AS SELECT * FROM Gold_Lead_Unsafe_Locations;

CREATE OR REPLACE TABLE Gold_Lead_Monthly_Avg
USING delta
AS SELECT * FROM Gold_Lead_Monthly_Avg;

CREATE OR REPLACE TABLE Gold_Lead_Exceedance_Events
USING delta
AS SELECT * FROM Gold_Lead_Exceedance_Events;

CREATE OR REPLACE TABLE Gold_Lead_Completeness_Report
USING delta
AS SELECT * FROM Gold_Lead_Completeness_Report;

CREATE OR REPLACE TABLE Gold_Lead_Risk_Summary
USING delta
AS SELECT * FROM Gold_Lead_Risk_Summary;
```
